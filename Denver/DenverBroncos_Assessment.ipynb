# Question 1: What is the name of the oldest dog in Southfield?

```python
#Finding the unique values in the Kind Column in Pets Dataframe
pets_df['Kind'].unique()
```

```python
#Renaming the column in pets DataFrame before merging
pets_df = pets_df.rename(columns={'Name':'PetName'})
```

```python
#Merging the two datasets using the OwnerID column as its common in both

merge_owner_pets_df = owner_df.merge(pets_df, on='OwnerID')
merge_owner_pets_df.head()
```

```python
#Finding results for dogs in Southfield City
dog_southfield = merge_owner_pets_df[(merge_owner_pets_df['City'] == 'Southfield') & (merge_owner_pets_df['Kind'] == 'Dog')]
dog_southfield
```

```python
#Checking the oldest dog
oldest_dog = max(dog_southfield['Age'])
oldest_dog
```

```python
#Detailed Result of Oldest Dog in Southfield
oldest_dog = dog_southfield.sort_values('Age', ascending=False).iloc[0]
oldest_dog
```

```python
#Printing the name of the Oldest Dog in Southfield
dogname_southfield = oldest_dog['PetName']
print(f'The name of the dog in Southfield city is: {dogname_southfield}')
```

# Question 3: Which owner spent the most on procedures for their pet(s)?

```python
#Merging all the datasets together
merged_df = merge_owner_pets_df.merge(procedures_df, on='PetID')
merged_df = merged_df.merge(proceduredetails_df, on=['ProcedureType','ProcedureSubCode'])
merged_df.head()
```

```python
#Checking the shape of the merged dataframe
merged_df.shape
```

```python
#Finding the result which owner spent the most on their pets
owner_spent = merged_df.groupby('OwnerID')['Price'].sum()
owner_mostspending = owner_spent.idxmax()
owner_spent
```

```python
#Resulting the owner who spent the most
owner_mostspent = merged_df.loc[merged_df['OwnerID'] == owner_mostspending]
owner_mostspent.head()
```

```python
#Printing the final answer which owner spent the most on their pet
owner_mostspent_name = owner_mostspent['Name'].iloc[0]
owner_mostspent_surname = owner_mostspent['Surname'].iloc[0]
print(f'The Owner who spent the most on their pets is: {owner_mostspent_name, owner_mostspent_surname}')
```

# Question 8: What percentage of cities have more male pets than female pets?

```python
#Segregating male and female pets by cities
male_pets_by_city = merge_owner_pets_df[merge_owner_pets_df['Gender'] == 'male'].groupby('City')['PetID'].count()
female_pets_by_city = merge_owner_pets_df[merge_owner_pets_df['Gender'] == 'female'].groupby('City')['PetID'].count()
```

```python
#Checking male pets by cities
male_pets_by_city
```

```python
#Checking female pets by cities
female_pets_by_city
```

```python
#Grouping the city and gender by petID
grouped_df = merge_owner_pets_df.groupby(['City', 'Gender'])['PetID'].count().reset_index()
grouped_df
```

```python
# Pivoting the data to get city as the index, gender as columns, and the count as values
pivot_df = grouped_df.pivot(index='City', columns='Gender', values='PetID').reset_index()
pivot_df.fillna(0, inplace=True)
pivot_df
```

```python
#Getting total number of cities
total_cities = owner_df['City'].nunique()
total_cities
```

```python
#Calculating the percentage of cities who have more male pets than female
pivot_df['more_male_pets'] = pivot_df['male'] > pivot_df['female']
cities_more_male_pets = (pivot_df['more_male_pets']).sum()
percentage_of_cities_have_more_male_pets = (cities_more_male_pets / total_cities)* 100
print(f'The percentage of cities who have more male pets than female : {percentage_of_cities_have_more_male_pets}')
```

# Question 9: Which city's pet sample is made up of exactly 70% dogs? The answer is case sensitive, so please match the value for City exactly.

```python
#Unstacking 'Kind' from owner and pets merged dataframe, grouping by City and Kind and storing it in pets by city
pets_by_city = merge_owner_pets_df.groupby('City')['Kind'].value_counts().unstack()
pets_by_city.fillna(0, inplace=True)
pets_by_city
```

```python
#Calculating percentage of dogs by city
pets_by_city['Percentage_of_dogs'] = (pets_by_city['Dog'] / (pets_by_city['Dog'] + pets_by_city['Cat'] + pets_by_city['Parrot'])) * 100
pets_by_city
```

```python
#Checking and Printing cities with exactly 70 percent of dogs
city_with_exact_70_percent_dog = pets_by_city[pets_by_city['Percentage_of_dogs'] == 70].index[0]
print(f'The cities with exactly 70 percent of dogs : {city_with_exact_70_percent_dog}')
```
