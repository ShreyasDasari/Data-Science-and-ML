# What is Support Vector Machine (SVM)?

Support Vector Machine, or SVM for short, is a potent machine learning technique used for regression and classification. A supervised learning technique called SVM can be used to categorize data based on a set of features. SVM is a binary classifier that functions by constructing a hyperplane, which is a line or plane that divides data into various classes.

By maximizing the margin between the two classes, SVM seeks to identify the optimum hyperplane that divides them. The margin is the separation between each class's closest points, known as support vectors, and the hyperplane. In order to maximize this margin and reduce classification error, SVM seeks out a hyperplane.

SVMs come in both linear and nonlinear varieties. The hyperplane in a linear SVM is a straight line that divides the data into two classes. The hyperplane in a nonlinear SVM can be a surface or a curved line that divides the data into various classes. In order to move the data into a higher-dimensional space where a linear hyperplane may be utilized to divide it, nonlinear SVMs employ a method known as the kernel trick.

Compared to other machine learning methods, SVM provides a number of benefits. Its capability to manage massive datasets and high-dimensional data is one of its primary features. SVM is resistant to outliers, or data points that differ dramatically from the rest of the data. Both linear and nonlinear data can be handled by SVM.

SVM, however, also has certain drawbacks. One of the key drawbacks is that the algorithm's performance might be impacted by the kernel function that is selected. SVM is slower and less scalable for really big datasets because it needs more processing and memory than other machine learning techniques.

In conclusion, SVM is a potent machine learning technique used for regression and classification. SVM divides data into classes using a hyperplane, optimizing the margin between classes while minimizing classification error. SVM is a flexible algorithm that can handle high-dimensional and nonlinear data, but it is more computationally and memory intensive than other algorithms and is sensitive to the kernel function selection.



# Support Vector Machines Terminology:

We must be familiar with some SVM terms:

**Hyperplane:**

A decision boundary known as a hyperplane divides a set of data points with several class labels. The greatest margin hyperplane is used by the SVM classifier to divide the data points into groups. This hyperplane is referred to as the maximum margin hyperplane, and the maximum margin classifier is the linear classifier it defines.


**Support Vectors:**

The sample data points that are closest to the hyperplane are called support vectors. By calculating margins, these data points will help define the separation line or hyperplane.

**Margin:**

The distance between the two lines on the closest data points is known as a margin. It is determined as the perpendicular distance between the line and the nearest data points or support vectors. We want to maximize this separation gap in SVMs to obtain the highest possible margin.

# Working of Support Vector Machine Algorithm:

* A well-liked machine learning algorithm called SVM (Support Vector Machines) is utilized for classification and regression applications. The fundamental goal of SVM is to identify the hyperplane that best discriminates between the various data point classes.

* The hyperplane is a line that separates the feature space into two sections, one for each class, in a binary classification problem. The margin—the distance between the hyperplane and the closest data points for each class—is maximized by the selection of the hyperplane.

* To accomplish this, SVM determines the margin by using the support vectors, or the data points that are most near the hyperplane. SVM aims to increase this margin while reducing classification error.

* If the data cannot be separated linearly, the SVM employs a kernel function to move the data into a higher-dimensional space where it can be separated linearly. The linear, polynomial, and radial basis function (RBF) kernels are the most frequently utilized kernel functions.

* SVM is a potent method that, in general, can achieve high accuracy in a variety of applications, including anomaly detection, text classification, and image classification.

# Mathematics behind SVM Algorithm for Classification and Regression:

**The Function of Decision:**

The following is a definition of the SVM's classification decision function:

$f(x) = \text{sign}(w^T x + b)$

where $w$ is the weight vector perpendicular to the hyperplane, $b$ is the bias term that pushes the hyperplane away from the origin, and $\text{sign}$ is the sign function that returns +1 for positive values and -1 for negative values. $x$ is the input feature vector.

**The Optimization Problem:**

Finding the ideal hyperplane that maximizes the margin between the classes while minimizing the classification error is the aim of SVM for classification. This is a problem that can be solved by optimization.

The equation for solving the SVM optimization problem can be written as follows:

**Minimize:**

$\frac{1}{2}||w||^2+C\sum_{i=1}^{N}\xi_{i}$

**Subject to:**

$y_{i}(w^{T}x_{i}+b) \geq 1-\xi_{i}$ and $\xi_{i} \geq 0$

where:

* $w$ is the weight vector
* $b$ is the bias term
* $C$ is a hyperparameter that controls the tradeoff between maximizing the margin and minimizing the classification error
* $N$ is the number of training examples
* $x_{i}$ is the feature vector of the i-th example
* $y_{i}$ is the label of the i-th example (+1 or -1)
* $\xi_{i}$ is a slack variable that allows for some misclassifications

# Kernel Trick:

When using SVMs, the kernel trick is a mathematical method for transforming the initial input data into a higher-dimensional feature space without actually computing the data's coordinates in that space. This enables SVMs to quickly identify a hyperplane that effectively separates the data with the greatest margin, solving non-linear classification issues.

SVMs can utilize a variety of kernel functions, however some of the more popular ones are as follows:

* **Linear Kernel:** For data that can be separated linearly, the linear kernel is utilized. It can be stated as follows:

$K(x, y) = x^T y$

where x and y are the input vectors.

Linear Kernel:

LinearKernel.PNG

The linear kernel is a straightforward, effective, and straightforward to compute kernel. In SVM classification, it is frequently employed as a baseline kernel, particularly when the data is high-dimensional. When the data cannot be separated linearly, however, it may not be successful; in this situation, a non-linear kernel may be utilized.

The linear kernel has the benefit of being less prone to overfitting than other kernel functions. This is due to the fact that it is less versatile and has fewer tuning parameters than non-linear kernels.

The interpretability of the linear kernel is another benefit. The hyperplane in the feature space that the linear kernel creates as the decision boundary is simple to see and comprehend.

In conclusion, the linear kernel is a straightforward and effective kernel that is applied when the data can be separated linearly. In comparison to non-linear kernels, it is less prone to overfitting and computes the dot product of two vectors to determine how similar they are. It is simple to interpret because of the hyperplane that serves as its decision boundary.


* **Polynomial Kernel:** The polynomial kernel uses a polynomial function to transfer the input data into a higher-dimensional feature space. It can be stated as follows:

$K(x, y) = (x^T y + c)^d$

where c is a constant and d is the degree of the polynomial.

Polynomial Kernel: 

PolynomialKernel.png

The decision boundary's complexity depends on the value of d. A decision boundary becomes more difficult as d is increased.

A constant term known as the parameter c controls how higher-order terms in the polynomial function affect the function. A decision boundary becomes more complex as c is greater because it emphasizes higher-order terms more. A lower value of c, on the other hand, emphasizes the lower-order terms more.

The polynomial kernel is helpful when the data may be split by a polynomial function but is not linearly separable in the original feature space. Think about a scenario where the data is arranged in a concentric circle, for instance. A linear classifier is incapable of classifying the data in the original feature space. The SVM may, however, identify a hyperplane that divides the data by translating the data to a higher-dimensional feature space using a polynomial function.

In SVM classification and regression, the polynomial kernel is a helpful tool for situations where the data cannot be divided linearly in the original feature space but can be separated by a polynomial function. The decision boundary's complexity and the impact of higher-order terms can both be adjusted using the parameters d and c.

* **Gaussian Kernel:** A Gaussian function is used by the Gaussian kernel to map the input data into an infinite-dimensional feature space. It can be stated as follows:

$K(x, y) = \exp(-\gamma \|x - y\|^2)$


where gamma is a parameter that controls the width of the Gaussian function.

Gaussian Kernel:

GaussianKernel.png

The radial basis function (RBF) kernel, usually referred to as the Gaussian kernel, is frequently employed in SVM classification and regression problems. When the data points in the input space cannot be separated linearly, the kernel is especially helpful.

The kernel function transforms the input data into a linearly separable infinite-dimensional feature space. This is accomplished by employing a Gaussian function to compute the pairwise similarity between data points in the input space.

The gamma () parameter of the Gaussian function regulates the kernel's width. A narrow Gaussian curve is produced by a small gamma value, and a larger Gaussian curve is produced by a large gamma value.In other words, a small gamma value suggests that the similarity drops gradually as the distance between two data points grows, whereas a large gamma value suggests that the similarity falls rapidly.

The ability to handle non-linearly separable data and the ability to capture complicated data structures are two benefits the Gaussian kernel offers over other kernels. It does, however, have some drawbacks, such as the kernel's sensitivity to gamma selection, which, if made carelessly, might result in overfitting. For large datasets, the kernel computation can also be computationally demanding.

* **Sigmoid Kernel:** A sigmoid function is used by the sigmoid kernel to transform the input data into a higher-dimensional feature space. In some machine learning applications, the Sigmoid kernel—which was initially created for use in neural networks—can be utilized as a neural network approximation. It can be stated as follows:

$K(x, y) = \tanh(-\alpha x^T y + c)$

where alpha and c are constants.

Sigmoid Kernel:

SigmoidKernel.png

Although the sigmoid kernel and the radial basis function (RBF) kernel are similar, they have different characteristics. One of the primary differences is that the sigmoid kernel has a finite range and is bounded. As a result, the sigmoid kernel is more suited for issues when the output values are constrained.

Classification and regression are two examples of machine learning tasks that can be performed using the sigmoid kernel. It is especially helpful in situations where a nonlinear decision boundary is needed and the data cannot be separated linearly.

The sigmoid kernel, however, has several shortcomings. The fact that it is sensitive to the selection of hyperparameters is one of the key problems. To guarantee good performance, the value of the parameter $-\alpha$ in particular needs to be carefully selected. Furthermore, overfitting of the sigmoid kernel can be reduced by employing regularization approaches.

The kernel function typically accepts two input vectors, x and y, and outputs a scalar value, K(x, y), that denotes how similar the two vectors are in the modified feature space. The SVM algorithm then employs this similarity metric to identify the hyperplane with the greatest margin of separation between the data.

In order to normalize or standardize the features of a dataset for machine learning, feature scaling is a crucial preprocessing step. The features of the training and testing datasets are scaled in the code above using the StandardScaler() method from the sklearn.preprocessing package.

Because many machine learning algorithms are sensitive to the scale of the input features, feature scaling is crucial. Larger magnitude features frequently outweigh smaller magnitude features, which might cause bias in the model. Furthermore, for some algorithms to function optimally the input characteristics must be normalized to have mean 0 and variance.

# **Calculating AUC Score and Plotting the ROC Curve**

A classifier model is used in binary classification to forecast the binary output (0 or 1) for a given input. Two well-known metrics—AUC score and ROC curve—are employed to assess the effectiveness of such a classifier model.

The performance of a binary classifier at various classification thresholds is shown graphically by the Receiver Operating Characteristic (ROC) curve. The True Positive Rate (TPR) and False Positive Rate (FPR) are represented on the y-axis and x-axis, respectively, of the ROC curve. The TPR measures the proportion of correctly categorized positive samples to all positive samples. The FPR measures the proportion of incorrectly labeled negative samples to all negative samples.

The binary classifier's performance at each potential classification threshold is summarized by the Area Under the Curve (AUC), a single numerical statistic. AUC scores range from 0 to 1, with a score of 0.5 denoting a guess at random and a score of 1 denoting a flawless classifier.

We must first forecast the likelihood that each sample in the test set will belong to the positive class before we can compute the AUC score and depict the ROC curve in SVM classification. Then, we use Scikit-learn's roc_curve() method to calculate the FPR, TPR, and thresholds at various probability thresholds. Finally, we can use the auc() method to visualize the ROC curve and calculate the AUC score.

The trade-off between TPR and FPR at various classification levels can be seen using the ROC curve and AUC score, which are significant evaluation measures. A model is deemed to be performing better if it has a higher AUC score and a higher TPR at a specific FPR.


```python
from sklearn.metrics import roc_auc_score,roc_curve

auc = roc_auc_score(y_test,y_pred)
print("AUC: %0.3f" %auc)
```

With an Area Under the Curve (AUC) score of 0.83, the SVM model performs well at differentiating between positive and negative samples. The score is a number between 0 and 1, where a score of 0.5 denotes a guess at random and a score of 1 denotes an ideal classifier. AUC = 0.83 indicates that the SVM model has a pretty good discriminatory power and is superior to a random guess.


```python
fpr, tpr, thresholds = roc_curve(y_test,y_pred)
```

```python
#Plotting the ROC Curve

plt.figure(figsize=(6,4))
plt.plot([0,1],[0,1],linestyle="--")
plt.plot(fpr,tpr, label='SVM (AUC = %0.2f)'% auc)
plt.xlabel("1-Specificity",fontsize=12)
plt.ylabel("Sensitivity",fontsize=12)
plt.legend(loc='lower right')
plt.show()
```

# **ROC Curve Explanation**

The ROC (Receiver Operating Characteristic) curve for an SVM (Support Vector Machine) model is plotted using the code above. The True Positive Rate (TPR) against False Positive Rate (FPR) for various categorization criteria is plotted on the ROC curve. The ROC curve is depicted using these predicted probabilities and the actual class labels after the SVM model predicts the likelihood of the positive class for each sample in the test set.

The FPR (1-Specificity) and TPR (Sensitivity) are plotted on the x- and y-axes, respectively. The top-left corner of the plot, which shows a perfect classifier and 100% TPR and 0% FPR, would be the optimum location. The SVM model performs better at correctly identifying positive data and avoids misclassification of negative samples the closer the SVM ROC curve is to the top-left corner. The SVM model performs well in the plot, with an **AUC of 0.83** showing that the model has a high ability to discriminate between positive and negative samples.



# **Implementing Model Cross Validation Using GridSearch**

```python
#Defining parameters for Hyperparameters Tuning
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
}

#Using the default SVC Kernel
svm = SVC(kernel='rbf',probability=True)

#Defining the gridsearch model
grid_search = GridSearchCV(svm, param_grid, verbose=2, cv=5)

#Fitting the model
grid_search.fit(X_train, y_train)

#Getting all the best scores
print('Best parameters:', grid_search.best_params_)
print('Best score:', grid_search.best_score_)
print('Best Estimator:', grid_search.best_estimator_)
```

```python
#Checking the predictions
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

grid_pred = grid_search.predict(X_test)
grid_pred
```

```python
#Calculating all the necessary scores required for analysis
print('Accuracy:', accuracy_score(y_test, grid_pred))
print('Precision:', precision_score(y_test, grid_pred))
print('Recall:', recall_score(y_test, grid_pred))
print('F1 score:', f1_score(y_test, grid_pred))
```

```python
#Reading the confusion matrix
print(confusion_matrix(y_test,grid_pred))
```

```python
#Getting the classification report
print(classification_report(y_test,grid_pred))
```

# **Calculating AUC Score and Plotting the ROC Curve**



```python
#Calculate AUC Score for GridSearchCV
auc_grid = roc_auc_score(y_test,grid_pred)
print('AUC: %.3f' % auc_grid)
```

# **Explaining Precision-Recall Graph**

The Precision-Recall curve, which shows a reasonable tradeoff between precision and recall, is pretty near to the top-right corner, as can be seen by looking at it. The model's average precision (AP), which is likewise displayed to be **0.72**, is a respectable number that indicates the model is functioning quite well.

Explaining the Precision-Recall Plot in detail:

A binary classification model's performance can be seen using a precision-recall (PR) graph, especially when the classes are unbalanced. The accuracy and recall values for various probability thresholds employed by the model to produce predictions are plotted on the PR graph.

Recall is the percentage of genuine positives among all real positives, whereas precision is the percentage of genuine positives among all projected positives. The occasions where the model predicted the positive class and the actual class was also positive are known as true positives (TP). False positives (FP) are situations where a positive class was predicted by the model but was really a negative class. False negatives (FN) occur when the model predicted a negative class, but the actual class was positive.

We first determine the precision and recall for various probability thresholds before creating a PR graph. Then, we graph these results with recall on the x-axis and precision on the y-axis. A particular threshold is represented by each point on the graph. This is how the PR curve can be understood.

1. The performance of the model improves as the curve gets closer to the top-right corner of the graph.
2. The accuracy-recall curve of a model with perfect precision and recall would begin at (0,1) and conclude at (1,1).
3. A model with a precision-recall curve that begins at (0,0) and finishes at (1,1) would exclusively predict positive examples.
4. A random model's precision-recall curve would resemble a diagonal from (0,0) to (1,1) in shape.

When the positive class is uncommon or when we wish to concentrate on the model's performance on the positive class, the PR graph is helpful. The appropriate trade-off between precision and recall can be determined using the PR graph, which can also assist in determining the probability threshold to employ when generating predictions.
